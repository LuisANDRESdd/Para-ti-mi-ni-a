**Inicialización de librerías y DataFrames**



```python
import pandas as pd
import numpy as np
from datetime import datetime

# DataFrame de Facturas (Usado para GroupBy)
facturas = pd.DataFrame({
    'folio': (15234, 15235, 15236, 15237, 15238, 15239, 15240, 15241, 15242),
    'sucursal': ('CDMX01', 'MTY01', 'CDMX02', 'CDMX02', 'MTY01', 'GDL01', 'CDMX02', 'MTY01', 'GDL01'),
    'monto': (1420.00, 1532.00, 890.00, 1300.00, 3121.47, 1100.5, 12230.00, 230.85, 1569.00),
    'fecha': (datetime(2019,3,11,17,24), datetime(2019,3,24,14,46), datetime(2019,3,25,17,58),
              datetime(2019,3,27,13,11), datetime(2019,3,31,10,25), datetime(2019,4,1,18,32),
              datetime(2019,4,3,11,43), datetime(2019,4,4,16,55), datetime(2019,4,5,12,59)),
    'cliente': (19234, 19232, 19235, 19233, 19236, 19237, 19232, 19233, 19232)
})

# DataFrame de Población (Usado para Limpieza de NaN)
poblacion = pd.DataFrame({
    'Animal': ('lobo', 'coyote', 'jaguar', 'cerdo salvaje', 'tapir', 'venado', 'ocelote', 'puma'),
    'Norte_I': (12, np.nan, None, 2, 4, 2, 14, 5),
    'Norte_II': (23, 4, 25, 21, 9, 121, 1, 2),
    'Centro_I': (15, 23, 2, 120, 40, 121, 0, 5),
    'Sur_I': (28, 46, 14, 156, 79, 12, 2, np.nan)
}).set_index('Animal')

# DataFrame secundario de otra_poblacion
otra_poblacion = pd.DataFrame({
    'Animal': ('lobo', 'coyote', 'jaguar', 'cerdo salvaje', 'tapir', 'venado', 'ocelote', 'puma'),
    'Norte_I': (12, 4, None, 2, 4, 2, 14, 4),
    'Norte_II': (23, 4, 25, 21, 9, 121, 1, 4),
    'Centro_I': (15, 4, 2, 120, 40, 121, 0, 4),
    'Sur_I': (28, 4, 14, 156, 79, 12, 2, 4)
}).set_index('Animal')
```

**PARTE 1: Métodos de Agrupamiento (groupby)**

Filtrado de columnas seleccionadas

```python
gastos_por_cliente = facturas.filter(items=["monto", "cliente"])
gastos_por_cliente
```

Creación del objeto groupby y visualización de grupos

```python
clientes = gastos_por_cliente.groupby("cliente")
clientes
```

Inspección de los índices de los grupos

```python
print(clientes.groups)
```

Iteración e impresión de grupos individuales

```python
for nombre_cliente, grupo in clientes:
    print(f"Grupo para cliente: {nombre_cliente}")
    print(grupo)
    print("-" * 50)
```
